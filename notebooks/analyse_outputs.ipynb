# Analyse complète des outputs

Ce notebook explore les principaux fichiers générés par le pipeline :

- **Topics LDA** (`lda_topics_*.xlsx`, `articles_*_avec_topics.xlsx`)
- **Clustering** (`articles_*_avec_clusters.xlsx`)
- **Classification SND30** (`objectifs_classifications_snd30_*.xlsx`)
- **Analyse budgétaire** (`analyse_budgetaire.xlsx`)
- **Tests statistiques** (`chi2_piliers.xlsx`, `test_mannwhitney.xlsx`)

L'objectif est de fournir une **vue d'ensemble reproductible** sur les résultats de l'audit sémantique.


```python
# Fonctions utilitaires

def describe_excel(path, n=5):
    """Charge un fichier Excel et affiche les premières lignes de chaque feuille.
    """
    path = Path(path)
    if not path.exists():
        print(f"⚠️ Fichier introuvable: {path}")
        return
    print(f"\n=== {path.name} ===")
    xls = pd.ExcelFile(path)
    for sheet in xls.sheet_names:
        print(f"\n--- Feuille: {sheet} ---")
        df = pd.read_excel(xls, sheet_name=sheet)
        display(df.head(n))
        print("Shape:", df.shape)

files
```

```python
# 1. Analyse budgétaire globale

analyse_budgetaire = files["analyse_budgetaire"]
describe_excel(analyse_budgetaire)

```

```python
# 2. Topics LDA et articles avec topics

for key in ["lda_topics_2024", "lda_topics_2025", "articles_2024_avec_topics", "articles_2025_avec_topics"]:
    describe_excel(files[key])

```

```python
# 3. Clustering des articles

for key in ["articles_2024_avec_clusters", "articles_2025_avec_clusters"]:
    describe_excel(files[key])

```

```python
# 4. Classification Zero-Shot SND30 des objectifs budgétaires

for key in ["objectifs_2024", "objectifs_2025"]:
    describe_excel(files[key])

```

```python
# 5. Tests statistiques (Chi², Mann-Whitney)

for key in ["chi2_piliers", "mannwhitney"]:
    describe_excel(files[key])

```
